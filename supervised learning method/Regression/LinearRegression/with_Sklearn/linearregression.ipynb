```python
import pandas as pd
import numpy as np 
import  matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/kaggle/input/housedata/house (2).csv')
data.head()
```

```python
data.info()
```

```python
plt.scatter(data['surface'], data['loyer'], color='blue', alpha=0.5)
plt.xlabel('Surface (m²)')
plt.ylabel('Loyer (€)')
plt.title('Relation entre la surface et le loyer')
plt.show()
```

```python
data['surface'].size
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train , y_test = train_test_split(data[['surface']], data['loyer'], test_size=0.2, random_state=42)
```

```python
model = LinearRegression().fit(X_train,y_train)
y_pred = model.predict(X_test)
x_vals = pd.DataFrame({'surface': range(int(y_pred.min()), int(y_pred.max())+1)})

print(y_pred)
```

```python
plt.plot(X_test, y_pred, color='red', label='Droite de régression')
plt.scatter(data['surface'], data['loyer'], color='blue', alpha=0.5)
plt.xlabel('Surface (m²)')
plt.ylabel('Loyer (€)')
plt.title('Relation entre la surface et le loyer')
plt.show()
```

```python
slope = model.coef_[0]         # θ₁ : the coefficient (slope)
intercept = model.intercept_   # θ₀ : the intercept (bias)
print(f"Estimated equation: y = {intercept:.2f} + {slope:.2f} * x")
```
